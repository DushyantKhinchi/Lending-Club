```
#Coverting string into datetime object
#This code is to convert a single string to datetime not a series
# from datetime import datetime
# datetime.strptime('Mar-2016-01', '%m-%y-%d')


#Pandas has its own function to convert string series to datetime object
lcdf['last_pymnt_d']=pd.to_datetime(lcdf['last_pymnt_d'],format='%b-%Y-%d',errors='coerce')
```

```
lcdf= lcdf[lcdf['last_pymnt_d'].notna()]
```

```
lcdf['last_pymnt_d'].isnull().sum()
```

```
lcdf.shape
```

```
lcdf['issue_d']=lcdf["issue_d"].str.replace("T00:00:00Z", "")
lcdf['issue_d']=pd.to_datetime(lcdf['issue_d'],format='%Y-%m-%d',errors='coerce')
```

```
from datetime import timedelta

lcdf['actual_term']=((lcdf['last_pymnt_d']-lcdf['issue_d']).dt.days)/365
lcdf['actual_term']
```

```
lcdf['actual_term'] = np.where(lcdf['loan_status'] != 'Fully Paid', 3.0, lcdf['actual_term'])
  
```

```
lcdf[{'loan_status','actual_term'}]
```

```
((lcdf['total_pymnt'] - lcdf['funded_amnt'])/lcdf['funded_amnt'])
```

```
lcdf['actual_return'] = ((lcdf['total_pymnt'] - lcdf['funded_amnt'])/lcdf['funded_amnt'])*(1/lcdf['actual_term'])*100
lcdf['actual_return']
```

```
#Derived attribute: proportion of satisfactory bankcard accounts 
lcdf['propSatisBankcardAccts']= np.where(lcdf['num_bc_tl']>0, lcdf['num_bc_sats']/lcdf['num_bc_tl'], 0)
lcdf['propSatisBankcardAccts']
```

```
#Another one - lets calculate the length of borrower's history with LC
#  i.e time between earliest_cr_line and issue_d

lcdf['earliest_cr_line']=lcdf['earliest_cr_line'].astype(str) + '-01'
lcdf['earliest_cr_line']=pd.to_datetime(lcdf['earliest_cr_line'],format='%b-%Y-%d')
lcdf['earliest_cr_line']
```

```
#Borrower's history
lcdf['borr_history']=((lcdf['issue_d']-lcdf['earliest_cr_line']).dt.days)/365
lcdf['borr_history']
```

```
#Another new attribute: ratio of openAccounts to totalAccounts
lcdf['openacc_totalacc_Ratio']=(lcdf['num_op_rev_tl']/lcdf['num_rev_accts'])
lcdf['openacc_totalacc_Ratio']
```

```
#removing features with more than 60% NaN values
lcdf = lcdf.loc[:, lcdf.isnull().mean() < .6]
```

```
#Removing Current loan status entry
index_Current = lcdf[lcdf['loan_status']=='Current'].index

lcdf.drop(index_Current, inplace=True)
```

```
#Converting loan status levels to '1'& '0'
lcdf['loan_status']=lcdf['loan_status'].astype('category').cat.codes
```

```
lcdf.drop(["actual_return","total_pymnt", "total_pymnt_inv", "last_credit_pull_d", "last_pymnt_amnt" , 'last_pymnt_d', 'recoveries', 'total_rec_int', 'total_rec_late_fee', 'debt_settlement_flag', 'out_prncp_inv', 'tot_cur_bal', 'num_il_tl', 'total_il_high_credit_limit', 'total_rec_prncp', 'collection_recovery_fee', 'out_prncp', 'avg_cur_bal', 'annRet','addr_state','zip_code','emp_title', 'actual_term','actual_return','out_prncp_inv','tot_cur_bal','num_il_tl','total_il_high_credit_limit','total_rec_prncp','collection_recovery_fee','out_prncp','avg_cur_bal'], axis=1, inplace=True)
```

```
#Columns with null values
null_columns = []
for column in lcdf:
    if lcdf[column].isnull().any():
       null_columns.append(column)
       print('Percentage of null values in {0} is {1} %'.format(column, round(lcdf[column].isnull().mean()*100,2)))
```

```
# num_tl_120dpd_2m and percent_bc_gt_75 has alomst 0 variance
lcdf.drop(['num_tl_120dpd_2m','percent_bc_gt_75'], axis=1,inplace=True)
```

```
lcdf['mths_since_last_delinq'] = lcdf['mths_since_last_delinq'].fillna((lcdf['mths_since_last_delinq'].median()))
lcdf['revol_util'] = lcdf['revol_util'].fillna((lcdf['revol_util'].median()))
lcdf['bc_open_to_buy'] = lcdf['bc_open_to_buy'].fillna((lcdf['bc_open_to_buy'].median()))
lcdf['bc_util'] = lcdf['bc_util'].fillna((lcdf['bc_util'].median()))
lcdf['mo_sin_old_il_acct'] = lcdf['mo_sin_old_il_acct'].fillna((lcdf['mo_sin_old_il_acct'].median()))
lcdf['mths_since_recent_bc'] = lcdf['mths_since_recent_bc'].fillna((lcdf['mths_since_recent_bc'].median()))
lcdf['mths_since_recent_inq'] = lcdf['mths_since_recent_inq'].fillna((lcdf['mths_since_recent_inq'].median()))
```

```
lcdf.drop(['emp_length'], axis=1,inplace=True)
```

```
#levels of purpose in X_train doesn't match X_test because X_test has an extra category 'educational'
#Removing 'educational' entry from X_test because it's an outlier

index_educational = lcdf[lcdf['purpose']=='educational'].index
lcdf.drop(index_educational, inplace=True)
```

```
y=lcdf['loan_status']
lcdf.drop(['loan_status'], axis=1, inplace=True)
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(lcdf, y, test_size=0.3, random_state=42)
```

```
X_train.drop({'earliest_cr_line','issue_d','pymnt_plan','title','application_type','disbursement_method','hardship_flag'}, axis=1, inplace=True)
X_test.drop({'earliest_cr_line','issue_d','pymnt_plan','title','application_type','disbursement_method','hardship_flag'}, axis=1, inplace=True)
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(X_train['term'].astype(str))
X_train['term']=le.transform(X_train['term'].astype(str))
X_test['term']=le.transform(X_test['term'].astype(str))

le.fit(X_train['grade'].astype(str))
X_train['grade']=le.transform(X_train['grade'].astype(str))
X_test['grade']=le.transform(X_test['grade'].astype(str))

le.fit(X_train['sub_grade'].astype(str))
X_train['sub_grade']=le.transform(X_train['sub_grade'].astype(str))
X_test['sub_grade']=le.transform(X_test['sub_grade'].astype(str))

le.fit(X_train['home_ownership'].astype(str))
X_train['home_ownership']=le.transform(X_train['home_ownership'].astype(str))
X_test['home_ownership']=le.transform(X_test['home_ownership'].astype(str))

le.fit(X_train['verification_status'].astype(str))
X_train['verification_status']=le.transform(X_train['verification_status'].astype(str))
X_test['verification_status']=le.transform(X_test['verification_status'].astype(str))

le.fit(X_train['purpose'].astype(str))
X_train['purpose']=le.transform(X_train['purpose'].astype(str))
X_test['purpose']=le.transform(X_test['purpose'].astype(str))

le.fit(X_train['initial_list_status'].astype(str))
X_train['initial_list_status']=le.transform(X_train['initial_list_status'].astype(str))
X_test['initial_list_status']=le.transform(X_test['initial_list_status'].astype(str))
```

```
X_train
```

```
## Import the random forest model.
from sklearn.ensemble import RandomForestClassifier 
rf = RandomForestClassifier(n_estimators=100) 
rf.fit(X_train, y_train)
```

```
rf.score(X_train, y_train)
```

```
rf.score(X_test, y_test)
```

```
import pandas as pd
feature_importances = pd.DataFrame(rf.feature_importances_,
                                   index = X_train.columns,
                                    columns=['importance']).sort_values('importance', ascending=False)
                                    
feature_importances.reset_index(inplace=True)
feature_importances
```

```
Top_30_features=feature_importances['index'][0:30]
```

```
Top_30_features
```

```
X_train=X_train[Top_30_features]
X_test=X_test[Top_30_features]
```

```
#Parameters used by our current forest

rf = RandomForestClassifier()
from pprint import pprint
print('Parameters currently in use:\n')
pprint(rf.get_params())
```

```
#Hyperparameter Tuning grid

from sklearn.model_selection import RandomizedSearchCV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 200, num = 5)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 50, num = 5)]
max_depth.append(None)

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]

# Method of selecting samples for training each tree
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```
from sklearn.model_selection import RandomizedSearchCV
# Using random grid to search for best hyperparameters
# base model
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 5, cv = 3, verbose=2, random_state=42, n_jobs = 1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```
#Best Parameters
rf_random.best_params_
```

```
#Optimized Random Forest
rf = RandomForestClassifier(n_estimators=125, bootstrap=True, max_depth=20, max_features='auto', min_samples_leaf=4, min_samples_split=2) 
rf.fit(X_train, y_train)
```

```
rf.score(X_train, y_train)
```

```
rf.score(X_test,y_test)
```

```
y_pred_test=rf.predict(X_test)
y_pred_train=rf.predict(X_train)
```

```
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_train,y_pred_train)

sns.heatmap(cm, square=True, annot=True, cbar=False, cmap='Wistia', linewidths=0.2, linecolor='k' )
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```
cm=confusion_matrix(y_test,y_pred_test)

sns.heatmap(cm, square=True, annot=True, cbar=False, cmap='Wistia', linewidths=0.2, linecolor='k' )
plt.xlabel('predicted value')
plt.ylabel('true value');
```

```
#ROC plot
import sklearn.metrics as metrics

# calculate the fpr and tpr for all thresholds of the classification
probs = rf.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc = metrics.auc(fpr, tpr)

# method I: plt
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
